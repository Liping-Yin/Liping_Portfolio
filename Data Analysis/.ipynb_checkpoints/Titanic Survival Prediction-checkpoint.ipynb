## 1.Data importing and processing

```python
# Importing Python Modules

#  Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Machine Learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier

import warnings
warnings.filterwarnings('ignore')

# configure graph styles
sns.set_theme(style = 'darkgrid')
plt.style.use('bmh')
plt.rcParams['figure.figsize'] = [6,3]
plt.rcParams['figure.dpi'] = 80

#import data
df_train = pd.read_csv('data/titanic/train.csv')
df_test = pd.read_csv('data/titanic/test.csv')

# check data information
df_train.head()


```

```python
# Information about the dataset
df_train.info()
```

### check missing values

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_train.isnull(),cbar=False, cmap='Blues')
plt.ioff()
```

As shown in the figure, lots of data of column **Cabin** is missing, this have to be dealt with before using the cabin data. In addition, some **Age** entries are missing too. A few values in **Embarked** is missing.



## Deal with Age Missing data


So, first check the statistical description of the data to get more insight

```python
df_train.describe()
```

Looking at the survied data, about only 38% people survied.(1: survived, 0: died)
The average age(30 years old),25 percentile is about 20, apparently if replace
missing value with mean value, the error will be increased. 

However, we can get different mean age according to passenger's attributes, noticing that in the **Name** column, passengers with different title, thus we can get the mean age of each group in the titles and replace the missing age with mean age values of different titles.

### Extract the title from Name

```python
# create new column named 'Title'
df_train['Title'] = 0
for i in df_train:
    df_train['Title'] = df_train.Name.str.extract('([A-Za-z]+)\.') # regular expression

#cross tab of title with gender(sex)
pd.crosstab(df_train.Title,df_train.Sex).T.style.background_gradient(cmap = 'Purples')

    
```

From the results, we can see, there are major titles: Master,Miss, Mr and Mrs. Some special title and similar meaning with the 4 major titles which can be replaced accordingly. 

    

```python
df_train.Title.replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess',
                        'Jonkheer','Col','Rev','Capt','Sir','Don'],
                       ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs',
                        'Other','Other','Other','Mr','Mr','Mr'],inplace = True)
pd.crosstab(df_train.Title,df_train.Sex).T.style.background_gradient(cmap = 'Purples')
```

### Mean Age based on Title

```python
df_train.groupby('Title')['Age'].mean()
```

### Replace Missing Age Values

```python
missing_values = df_train.Age.isnull()
titles = df_train.Title
df_train.loc[missing_values & (titles == 'Mr'),'Age'] = 33
df_train.loc[missing_values & (titles == 'Mrs'),'Age'] = 36
df_train.loc[missing_values & (titles == 'Master'),'Age'] = 5
df_train.loc[missing_values & (titles == 'Miss'),'Age'] = 22
df_train.loc[missing_values & (titles == 'Other'),'Age'] = 46

# check if replacing missing values is done
missing_values.any()
```

## Deal with Cabin Missing data

too much missing data to do something useful.In this case, either remove the cabin column or give tags like 0/1 to the values.
Here I prefer to replace NaN/not NaN with 0/1

```python

df_train['Cabin'] = df_train['Cabin'].apply(lambda x: 1 if isinstance(x,str) else 0)

sns.countplot(x = 'Cabin',data = df_train)
```

## Deal with Embarked Missing Value

```python
df_train['Embarked'] = df_train['Embarked'].fillna('S') # assmue one Embark location
df_train.info()
```

Now, all the missing values ware handled.

Next, in order to use machine learning, all the categorical data should be converted to dummy variables.

**Map the categorical data**

```python
train = df_train.copy()

# Mapping Sex
sex_map = {'female':0,'male':1}
train['Sex'] = train['Sex'].map(sex_map).astype(int)

# Mapping title
title_map = {'Master':0,'Miss':1,'Mr':2,'Mrs':3,'Other':4}
train['Title'] = train['Title'].map(title_map).astype(int)

# Mapping Embarked

embark_map = {'S':0,'C':1,'Q':2}
train['Embarked'] =  train['Embarked'].map(embark_map).astype(int)

# Alternative way to map categorical data 

# features = ['Pclass','Sex','SibSp','Parch','Title','Embarked']
# train_dum = pd.get_dummies(df_train[features])
# train.drop(df_train[featuresres].axis=1,intplace = True)
# train = pd.concat([train,train_dum],axis=1)
```

```python
train.head()
```

Since Name mainly relates with  sex or age, so Name column is dependent on other columns and hence not useful. And the ticket is composed of random strings , so it is not so useful feature either.
So, when we analysis the data, we can drop those two columns.


```python
# drop Name and Ticket columns
train = train.drop('Name',axis = 1 )
train = train.drop('Ticket',axis = 1)
```

```python
train.describe(include='all')
```

### Deal with test data in the same way as train data

```python
# Deal with Missing Age
df_test['Title'] = 0
for i in df_test:
    df_test['Title'] = df_test.Name.str.extract('([A-Za-z]+)\.')

df_test.Title.replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess',
                        'Jonkheer','Col','Rev','Capt','Sir','Dona'],
                       ['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs',
                        'Other','Other','Other','Mr','Mr','Miss'],inplace = True)

df_test.groupby('Title')['Age'].mean()

    
missing_values = df_test.Age.isnull()
titles = df_test.Title
df_test.loc[missing_values & (titles == 'Mr'),'Age'] = 32
df_test.loc[missing_values & (titles == 'Mrs'),'Age'] = 39
df_test.loc[missing_values & (titles == 'Master'),'Age'] = 7
df_test.loc[missing_values & (titles == 'Miss'),'Age'] = 22
df_test.loc[missing_values & (titles == 'Other'),'Age'] = 43

# Deal with Cabin Missing values
df_test['Cabin'] = df_test['Cabin'].apply(lambda x: 1 if isinstance(x,str) else 0)
# Deal with Fare Missing value
df_test['Fare'] = df_test['Fare'].fillna(0)

# Mapping categorial data
test = df_test.copy()

# Mapping Sex
sex_map = {'female':0,'male':1}
test['Sex'] = test['Sex'].map(sex_map).astype(int)

# Mapping title
title_map = {'Master':0,'Miss':1,'Mr':2,'Mrs':3,'Other':4}
test['Title'] = test['Title'].map(title_map).astype(int)

# Mapping Embarked

embark_map = {'S':0,'C':1,'Q':2}
test['Embarked'] =  test['Embarked'].map(embark_map).astype(int)
# Drop Name and Ticket columns
test = test.drop('Name',axis = 1 )
test = test.drop('Ticket',axis = 1)

print(test.info())
    
```

Now, the cleaned dataset is ready for machine learning
